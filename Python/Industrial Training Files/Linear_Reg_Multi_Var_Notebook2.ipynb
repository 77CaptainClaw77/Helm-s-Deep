```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
```

```python
mydata=pd.read_csv('data2.csv',names=['Area','Bedrooms','Price'])
```

```python
mydata
```

```python
data_set=(mydata-mydata.mean())/mydata.std()
```

```python
data_set
```

```python
X=data_set.iloc[:,0:2]
```

```python
X.insert(0,'Ones',1)
```

```python
X
```

```python
y=data_set.iloc[:,2:3]
```

```python
y
```

```python
def ComputeCost(X,y,theta):
    return np.sum(np.power(((X*theta.T)-y),2))/ (2*len(X))
```

```python
X=np.matrix(X)
y=np.matrix(y)
theta=np.matrix(np.array([0,0,0]))
```

```python
X.shape
```

```python
theta.shape
```

```python
ComputeCost(X,y,theta)
```

```python
def gradientDescent(X,y,theta,alpha,iterations):
    temp=np.matrix(np.zeros(theta.shape))
    parameter=3
    cost=np.zeros(iterations)
    for i in range(iterations):
        error=(X*theta.T)-y
        for j in range(parameter):
            term=np.multiply(error,X[:,j])
            temp[0,j]=theta[0,j]-(alpha/len(X))*np.sum(term)
        theta=temp
        cost[i]=ComputeCost(X,y,theta)
    return theta,cost
```

```python
a,b=gradientDescent(X,y,theta,0.01,2000)
```

```python
a
```

```python
b[-4]
```

```python
x=np.linspace(data_set.Area.min(),data_set.Area.max(),100)
from mpl_toolkits.mplot3d import Axes3D
x_a=np.linspace(data_set.Bedrooms.min(),data_set.Bedrooms.max(),100)
f=a[0,0]+a[0,1]*x+a[0,2]*x_a
#fig,ax=plt.subplots(figsize=(12,8),projection='3d')
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,x_a,f)
ax.scatter(data_set.Area,data_set.Bedrooms,data_set.Price)
```
